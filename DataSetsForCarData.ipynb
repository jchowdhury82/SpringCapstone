```python
# Get car reliability index

url = 'https://www.reliabilityindex.com/manufacturer/relIndex'
response = requests.get(url).content.decode('utf-8')
dfs = pd.read_html(response) 
dfrel = dfs[0]
dfrel.drop(columns = ['Reliability Index'],inplace = True)
dfrel = dfrel.rename({'Make and Model' : 'Brand','Reliability Index.1' : 'Reliability Index'},axis = 1)
dfrel
```
